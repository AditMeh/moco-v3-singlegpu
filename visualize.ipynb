```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
img = "/home/aditmeh/Desktop/moco-v3-singlegpu/imagenette2/train/n01440764/ILSVRC2012_val_00016018.JPEG"

img = Image.open(img)
```

```python
transform = transforms.Compose([
    transforms.RandomResizedCrop((16, 16))
])

# Prepare to plot a 5x5 grid
fig, axs = plt.subplots(5, 5, figsize=(12, 12))
for i in range(5):
    for j in range(5):
        # Apply the random crop transform
        cropped_img = transform(img)
        # Convert to numpy for plotting
        axs[i, j].imshow(np.array(cropped_img))
        axs[i, j].axis('off')

plt.tight_layout()
plt.show()
```
